The Employee Management System is software developed in python programming language and it is a desktop application. This system is created using tkinter library. This Management System has MySQL database 

An Employee MANAGEMENT system is a software or application which is used to manage employee information/data. This system is very useful for companies, colleges 
Employee Management System is designed to track, maintain and manage all the data generated by a companies, We are storing employee information in database like name, mobile number ,email id, address, department, salary , here We are doing various operation ,like add ,delete ,update, search, show result and other operations.

```python
import mysql.connector
from tkinter import *
from tkinter import messagebox
from tkinter import ttk

win=Tk()
win.title('EMPLOYEE MANAGEMENT SYSTEM')
win.config(bg='skyblue')
win.geometry('800x600')

def show():
    db=mysql.connector.connect(user='root',password='Python@123',host='localhost',database='emp_project')
    qur='SELECT * FROM emp_table'
    mycur=db.cursor()
    mycur.execute(qur)
    list1=mycur.fetchall()
    for i in list1:
        treev.insert('',END,values=(i[0],i[1],i[2],i[3]))
    mycur.close()
    db.close()
    


def add():
    name=n1.get()
    mobno=m1.get()
    dept=d1.get()
    salary=s1.get()
    if name=='' or mobno=='' or dept=='' or salary=='':
        messagebox.showinfo('info','All fields are compulsory')
    else:
        db=mysql.connector.connect(user='root',password='Python@123',host='localhost',database='emp_project')
        qur=f'INSERT INTO emp_table VALUES("{name}","{mobno}","{dept}","{salary}")'
        mycur=db.cursor()
        mycur.execute(qur)
        db.commit()
        mycur.close()
        db.close()
        messagebox.showinfo('info','Data is inserted successfully')
        n1.set('')
        m1.set('')
        d1.set('')
        s1.set('')


def salary():
    win.destroy()
    root=Tk()
    root.title('Salary')
    root.configure(bg='cyan')
    root.geometry('800x800')
    def fun():
        sal_name=var3.get()
        db=mysql.connector.connect(user='root',password='Python@123',host='localhost',database='emp_project')
        qur=f'SELECT salary FROM emp_table WHERE name="{sal_name}"'
        mycur=db.cursor()
        mycur.execute(qur)
        tup2=mycur.fetchone()
        l8.config(text=tup2[0]+' Rs')
        mycur.close()
        db.close()
    
    l7=Label(root,text='ENTER NAME',bd=5,relief='ridge',width=20)
    l7.place(x=100,y=40)
    var3=StringVar()
    e7=Entry(root,textvariable=var3,width=40,bd=5,relief='ridge')
    e7.place(x=260,y=40)
    b9=Button(root,text='SHOW',width=20,command=fun)
    b9.place(x=260,y=80)
    l8=Label(root,text='Salary is here',height=4,width=25)
    l8.place(x=100,y=150)
    
    root.mainloop()


def delete():
    del_name=var1.get()
    if del_name=='':
        messagebox.showinfo('info','All fields are compulsory')
    else:
        db=mysql.connector.connect(user='root',password='Python@123',host='localhost',database='emp_project')
        qur=f'DELETE FROM emp_table WHERE name="{del_name}"'
        mycur=db.cursor()
        mycur.execute(qur)
        db.commit()
        mycur.close()
        db.close()
        messagebox.showinfo('info','Data is deleted successfully')
        var1.set('')

def select():
    up_name=var2.get()
    if up_name=='':
        messagebox.showinfo('info','All fields are compulsory')
    else:
        db=mysql.connector.connect(user='root',password='Python@123',host='localhost',database='emp_project')
        qur=f'SELECT * FROM emp_table WHERE name="{up_name}"'
        mycur=db.cursor()
        mycur.execute(qur)
        tup1=mycur.fetchone()
        e1.insert(0,tup1[0])
        e2.insert(0,tup1[1])
        e3.insert(0,tup1[2])
        e4.insert(0,tup1[3])
        mycur.close()
        db.close()
    

def update():
    name=n1.get()
    mobno=m1.get()
    dept=d1.get()
    salary=s1.get()
    if name=='' or mobno=='' or dept=='' or salary=='':
        messagebox.showinfo('info','All fields are compulsory')
    else:
        db=mysql.connector.connect(user='root',password='Python@123',host='localhost',database='emp_project')
        qur=f'UPDATE emp_table SET mobno="{mobno}",dept="{dept}",salary="{salary}" WHERE name="{name}"'
        mycur=db.cursor()
        mycur.execute(qur)
        db.commit()
        mycur.close()
        db.close()
        messagebox.showinfo('info','Data is updated successfully')
        n1.set('')
        m1.set('')
        d1.set('')
        s1.set('')


def clear():
    treev.delete(*treev.get_children())

# main label

l0=Label(win,text='EMPLOYEE MANAGEMENT SYSTEM',bd=5,relief='ridge',bg='white',fg='black',
         width=40,font=('time new roman',16,'bold'))

l0.place(x=268,y=40)


# name >> label entry

l1=Label(win,text='EMPLOYEE NAME',bd=5,relief='ridge',bg='white',fg='black',
         width=20,font=('time new roman',12,'bold'))

l1.place(x=190,y=100)

n1=StringVar()
e1=Entry(win,textvariable=n1,bd=5,relief='ridge',bg='white',fg='black',
         width=32,font=('time new roman',12,'bold'))
e1.place(x=405,y=100)

# mobile number >> label entry

l2=Label(win,text='MOBILE NUMBER',bd=5,relief='ridge',bg='white',fg='black',
         width=20,font=('time new roman',12,'bold'))

l2.place(x=190,y=140)

m1=StringVar()
e2=Entry(win,textvariable=m1,bd=5,relief='ridge',bg='white',fg='black',
         width=32,font=('time new roman',12,'bold'))
e2.place(x=405,y=140)


# dept > label entry

l3=Label(win,text='DEPARTMENT',bd=5,relief='ridge',bg='white',fg='black',
         width=20,font=('time new roman',12,'bold'))

l3.place(x=190,y=180)

d1=StringVar()
e3=Entry(win,textvariable=d1,bd=5,relief='ridge',bg='white',fg='black',
         width=32,font=('time new roman',12,'bold'))
e3.place(x=405,y=180)

# salary > label entry

l4=Label(win,text='SALARY',bd=5,relief='ridge',bg='white',fg='black',
         width=20,font=('time new roman',12,'bold'))

l4.place(x=190,y=220)

s1=StringVar()
e4=Entry(win,textvariable=s1,bd=5,relief='ridge',bg='white',fg='black',
         width=32,font=('time new roman',12,'bold'))
e4.place(x=405,y=220)

# 4 buttons...ADD,SHOW,SALARY,EXIT

b1=Button(win,text='SHOW',width=20,command=show)
b1.place(x=200,y=320)

b2=Button(win,text='ADD',width=20,command=add)
b2.place(x=500,y=320)

b3=Button(win,text='EXIT',width=20,command=win.destroy)
b3.place(x=200,y=360)

b4=Button(win,text='SALARY',width=20,command=salary)
b4.place(x=500,y=360)

# DELETE >>> label,entry,button

l5=Label(win,text='ENTER THE NAME>',bd=5,relief='ridge',bg='white',fg='black',
         width=20,font=('time new roman',12,'bold'))

l5.place(x=190,y=400)

var1=StringVar()
e5=Entry(win,textvariable=var1,bd=5,relief='ridge',bg='white',fg='black',
         width=32,font=('time new roman',12,'bold'))
e5.place(x=405,y=400)


b5=Button(win,text='DELETE',width=20,command=delete)
b5.place(x=385,y=440)

# update >> label,entry,2 buttons

l6=Label(win,text='ENTER THE NAME>',bd=5,relief='ridge',
         bg='white',fg='black',width=20,font=('times new roman',12,'bold'))
l6.place(x=190,y=480)

var2=StringVar()
e6=Entry(win,textvariable=var2,bd=5,relief='ridge',
         bg='white',fg='black',width=38,font=('times new roman',12,'bold'))
e6.place(x=385,y=480)

b6=Button(win,text='SELECT',width=20,command=select)
b6.place(x=230,y=520)
b7=Button(win,text='UPDATE',width=20,command=update)
b7.place(x=385,y=520)


# treeview

treev=ttk.Treeview(win,height=23)
treev.place(x=760,y=100,width=500)

treev["columns"] = ("1", "2", "3","4")
treev['show'] = 'headings'


treev.column("1", width = 90, anchor ='c')
treev.column("2", width = 90, anchor ='se')
treev.column("3", width = 90, anchor ='se')
treev.column("4", width = 90, anchor ='se')

treev.heading("1", text ="Name")
treev.heading("2", text ="Mobile No")
treev.heading("3", text ="Dept")
treev.heading("4", text ="Salary")

# clear button

b8=Button(win,text='CLEAR',width=20,command=clear)
b8.place(x=950,y=590)



win.mainloop()
```
